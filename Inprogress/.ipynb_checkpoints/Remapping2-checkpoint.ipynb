```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
from scipy.interpolate import griddata
from pylab import *
from multiprocessing import Pool
```

```python
Comp = sys.argv[1]
OP = sys.argv[2]
Years = sys.argv[3]
```

```python
def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
```

```python
Comp="Mobilis"
OP="3day"
Years="200-220"
x="/noc/msm/scratch/students/hb1g13/"+Comp+"/"+OP
os.chdir(x)
```

```python
Rho = np.genfromtxt('Temp', delimiter = ',') 
Rho=Rho[0:42]
```

```python
file = netcdf.NetCDFFile(Years+"Psi.nc",'r')
lvrho=file.variables['LaVH1TH']
lvrho=lvrho[:]
Y=file.variables['Yp1']
Y=Y[:]
time=file.variables['T']
ti=time[:]
dx=Y[1]-Y[0]
```

```python
VT=np.sum(lvrho*dx,axis=3)
VTf=VT[:,::-1,:]
```

```python
VTfdz=np.cumsum(VTf,axis=1)
```

```python
VTdx=VTfdz[:,::-1,:]
psi=np.mean(VTdx,axis=0)/10**6
y=Y/1000
start=int(np.divide(ti[0],(86400*360)))
end=int(np.divide(ti[-1],(86400*360)))
```

```python
plt.contourf(y,Rho,psi,50,cmap=cm.seismic)
cbar = plt.colorbar()
clim((-np.nanmax(abs(psi)),np.nanmax(abs(psi))))
plt.title("RMOC for years "+str(start)+"-"+str(end)+" ")
plt.xlabel('Distance (km)')
plt.ylabel('Temperature $^o$C')
cbar.ax.set_ylabel("$\psi \,\, (sv)$")
#plt.text(1500, 4, r"AAIW", fontsize=17, color="white")
#plt.text(1500, 1.5, r"NADW", fontsize=16, color="white")
#plt.text(1500, 0, r"AABW", fontsize=16, color="white")
x="/noc/users/hb1g13/Figures/"+Comp+"/"+OP
if not os.path.exists(x):
       os.makedirs(x)
plt.savefig(x+"/Psires"+str(start)+"-"+str(end)+"(annotated).png")
```

```python
NA=np.sum(psi[psi>0]) #Average strength of NADW Cell
psibw=psi[0:22,:]
AB=np.sum(psibw[psibw<0]) #Av
psiiw=psi[20::-1,:]
AI=np.sum(psiiw[psiiw<0]) #Av
NAav=np.mean(psi[psi>0.2]) #Average strength of NADW Cell
ABav=np.mean(psibw[psibw<-0.2]) #Av
AIav=np.mean(psiiw[psiiw<-0.2]) #Av
TotalRMOC=np.sum(abs(psi))
file = open("/noc/users/hb1g13/Figures/"+Comp+"/"+OP+"Transports.txt", "w")
l1="The total absolute integrated transport is "+str(TotalRMOC)
l2="The total NADW Transport is "+str(NA)
l3= "The total AABW Transport is "+str(AB)
l4="The total AAIW Transport is "+str(AI)
l5="The mean NADW Transport is "+str(NAav)
l6="The mean AABW Transport is "+str(ABav)
l7="The mean AAIW Transport is "+str(AIav)
nl = "\n"
file.writelines([l1,nl,l2,nl,l3,nl,l4,nl,l5,nl,l6,nl,l7])
file.close()
```

```python
file2 = netcdf.NetCDFFile(Years+"all.nc",'r')
Temp=file2.variables['THETA']
Temp=Temp[:]
U=file2.variables['UVEL']
U=U[:]
```

```python
Psimap=np.zeros(shape(Tavlat))
```

```python
for i in range(len(Yc)):
    for k in range(len(Z)):
        #D=Densav[k,i]
        D=Tavlat[k,i]
        if D==nan:
           Psimap[k,i]=nan
        else:
            P=psi[:,i]
            I=find_nearest(Rho, D)
            b=nonzero(Rho==I)[0][0]
            Psimap[k,i]=P[b]
```

```python
lmav=np.mean(lmc,axis=2)
```

```python
Psimapped=np.multiply(Psimap,lmav)
Psimapped[Psimapped==0]=np.nan
```

```python
plot1=plt.contourf(Yc,Z,Psimapped,150,cmap=cm.seismic)
cbar = plt.colorbar()
clim(-(np.nanmax(abs(psi))),np.nanmax(abs(psi)))
plt.title('Remapped Stream Function')
plt.xlabel('Distance (km)')
plt.ylabel('Depth')
cbar.ax.set_ylabel('$\psi_{res}$ (Sv)')
x="/noc/users/hb1g13/Figures/"+Comp+"/"+OP 
if not os.path.exists(x):
       os.makedirs(x)
y=x+"/Reampped"+Years+".png" 
plt.savefig(y)
```
