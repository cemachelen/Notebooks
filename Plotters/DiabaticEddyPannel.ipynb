```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
import glob
from scipy.interpolate import interp1d
from scipy import interpolate
from numba import autojit
from pylab import *
```

```python
check=0
runs=[]
for i in range(len(tau)):
    flist=path+'/'+str(tau[i])+'daynokpp/'+Year
    if not os.path.exists(flist):
       print 'WARNING: '+flist+' does not exist! (skipping this ta\
u...)'
       check+=0
    else:
       check+=1
       runs.append(i)
```

```python
def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
def regrid(Variable):
    Vc=(Variable[:,0:-1]+Variable[:,1::])/2
    return Vc
numba_regrid = autojit()(regrid)
numba_regrid.func_name = "numba_regrid"
```

```python
Runs=np.array(runs)
```

```python
# Constant Parameters
alpha=2.000000E-04
G=-9.81
rho_0=1000
C_p=3985
Res=5000
Ly=2000e3
Lx=1000e3 
H=2985 
nz=30 
#Dimensions
nx=np.round(Lx/Res)
ny=np.round(Ly/Res)
dx=np.ones(nx)*Res
dy=np.ones(ny)*Res
x=(np.cumsum(dx)-dx/2)-Lx/2
y=(np.cumsum(dy)-dy/2)-Ly/2
[Y, X]=np.meshgrid(y,x) 
Q=10*(np.sin(Y*(3*pi/Ly)))
Q[:,ny-(np.round(ny/6)):ny]=0
Q=np.zeros(np.shape(Q))
Q[:,0:int(5*ny/36)]=-11*(np.sin(Y[:,0:int(5*ny/36)]*(pi/Y[1,int(5*ny/36)])))
Q[:,int(5*ny/36):int(20*ny/36)]=10*(np.sin((Y[:,int(5*ny/36):int(20*ny/36)]-Y[1,int(5*ny/36)])*(pi/(5*Y[1,int(15*ny/36)]))))
Q[:,int(20*ny/36):int(30*ny/36)]=-10.5*(np.sin((Y[:,int(20*ny/36):int(30*ny/36)]-Y[1,int(20*ny/36)])*(0.8*pi/(Y[1,int(10*ny/36)]))))
Q=Q+(-sum(Q)/(ny*nx)) 
Psipred=(-G*alpha/(rho_0*C_p))*-(Q[1,:])
```

```python
# Grid parameters
gridfilepath=path+'/'+str(tau[0])+'daynokpp/'
file2=netcdf.netcdf_file(gridfilepath+'grid.nc','r')
Zp1=file2.variables['Zp1']
Zp=Zp1[:]*1
Z=file2.variables['Z']
Z=Z[:]*1
Y=file2.variables['Yp1']
Y=Y[:]*1
Yc=file2.variables['Y']
Yc=Yc[:]*1
```

```python
fig=plt.figure(figsize=(12.5,12.5))
for i in range(len(Runs)):
    fname=path+'/'+str(tau[Runs[i]])+'daynokpp/'+Year
    file2read = netcdf.NetCDFFile(fname,'r')
    lvrho=file2read.variables["LaVH1TH"]
    lvrho=lvrho[:]
    time=file2read.variables['T']
    ti=time[:]
    dx=Y[1]-Y[0] # Find Resolution
    VT=np.sum(lvrho*dx,axis=3) #integrate Vdx along x
    VTfdz=np.cumsum(VT[:,::-1,:],axis=1) #sum up the water column
    psi=np.mean(VTfdz[:,::-1,:],axis=0)/10**6 #Time average and put into Sv and put back in right order
    z=np.array(range(1,60))
    Rho = np.genfromtxt(path+'/'+str(tau[Runs[i]])+'daynokpp/Temp', delimiter = ',')
    nolayers=len(psi[:,1])
    Rho=Rho[0:nolayers]#The layers package bins a layer so adjust \
    start=int(np.divide(ti[0],(86400*360)))#Find run start and sto\
    end=int(np.divide(ti[-1],(86400*360)))
    a=list(fname)
    a[-6]='a'
    a[-5]='l'
    a[-4]='l'
    A=''.join(a)
    if not os.path.isfile(A):
       print "Warning: file "+A+" not found, so i'm skipping this!"
       continue
    file2 = netcdf.NetCDFFile(A,'r')
    Temp=file2.variables['THETA']
    Temp=Temp[:]*1
    Tav=np.mean(Temp,axis=0)
    Tavlat=np.mean(Tav,axis=2)
    dz=Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
    Psi=numba_regrid(psi)
    #Remap Part                                                                                       
    #Expand temperature co-ordinates (30 lvls to 168 lvls)                                            
    Z2=interp1d(Z,Z,axis=0)
    Znew=linspace(int(Z[0]),int(Z[-1]),168)
    Zexp=Z2(Znew)
    T2=interp1d(Z,Tavlat,axis=0)
    Tnew=linspace(int(Z[0]),int(Z[-1]),168)
    Texp=T2(Tnew)
    R2=interp1d(Rho,Rho,axis=0)
    Rnew=linspace(Rho[0],Rho[-1],168)
    Rexp=R2(Rnew)
    P2=interp1d(Rho,psi,axis=0)
    Pnew=linspace(Rho[0],Rho[-1],168)
    Pexp=P2(Pnew)
    Psimap=np.zeros(shape(Texp))
    for g in range(len(Yc)):
        for k in range(len(Zexp)):
            D=Texp[k,g]
            if np.isnan(D):
               Psimap[k,g]=np.nan
            else:
               P=Pexp[:,g]
               I=find_nearest(Rexp, D)
               b=nonzero(Rexp==I)[0][0]
               Psimap[k,g]=P[b]
    B=np.zeros(np.shape(Yc[:]))
    db=np.zeros(np.shape(Yc[:]))
    Tavlatav=np.nanmean(Tavlat[0:12,:],axis=0)
    A=np.gradient(Tavlatav,1)
    B=A*alpha*G/5000
    C=np.mean(abs(B[10:-10]))
    #B=np.mean(-G*alpha*Texp[0:12,:],axis=0)
    Psi_isoB=np.zeros(np.shape(Yc[:]))
    Psi_isoB[3:-3]=B[3:-3]-C
    Psi_isoB2=-Psimap[12,:]*Psi_isoB
    D=Psi_isoB2-Psipred
    #Psimapped=np.multiply(Psimap,lmav) HFacS is req for topo runs                                    
    #Psimap[Psimap==0]=np.nan NP can't do this!! 10000 yr has 0 residul
    #Now put the MOC into 168 lvls to make the eddy plot     
    if Full=='N':
         ax = fig.add_subplot(2, 2, i+1)
    else:
         ax = fig.add_subplot(3, 3, i+1)
    if VAR=='D':
        p=ax.plot(Yc/1000,D)
        s=ax.plot(Yc/1000,Psi_isoB2,'k')
        r=ax.plot(Yc/1000,(Psipred),'r')
        #s=ax.plot(Yc/1000,Psi_isoB2,'g')
    else:
        p=ax.plot(Yc/1000,D)
        s=ax.plot(Yc/1000,Psi_isoB2,'k')
        r=ax.plot(Yc/1000,(Psipred),'r')
    ax.set_title(str(tau[Runs[i]])+'day')
    if str(tau[Runs[i]])=='Closed':
        ax.set_title(str(tau[Runs[i]]))
    ax.set_xlabel('Distance (km)')
    ax.set_ylabel('$m^2s^{-3}$')
plt.tight_layout()
#lgd=legend([p,s,r],[r'Diabatic','$\psi_{iso}$','B'],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
lgd=plt.legend([r'Diabatic','$\psi_{iso}b_y$','B'], shadow=True,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#
q=( os.path.expanduser('~')+"/Figures/Mobilis")
if not os.path.exists(q):
       os.makedirs(q)
#plt.figlegend( [p,s,r],[r'Diabatic','$\psi_{iso}$','B'], loc = 'lower center', ncol=5, labelspacing=0. )
if VAR=='K':
    plt.savefig(q+"/Diabaticpanel"+str(start)+"-"+str(end)+".png",bbox_extra_artists=(lgd,), bbox_inches='tight')
else:
    plt.savefig(q+"/Diabaticpanel"+str(start)+"-"+str(end)+".png")
```
