```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
import glob
import matplotlib.gridspec as gridspec
```

```python
#tau=['3','10','30','100','300','1000','3000','10000','Closed']
# Not all plots - too many lines!
tau=['3','30','300','3000','Closed']
x='/noc/msm/scratch/students/hb1g13/Mobilis'
```

```python
check=0
runs=[]
for i in range(len(tau)):
    flist=x+'/'+str(tau[i])+'daynokpp/VTprimebar.nc'
    if not os.path.exists(flist):
       print 'WARNING: '+flist+' does not exist! (skipping this ta\
u...)'
       check+=0
    else:
       check+=1
       runs.append(i)
```

```python
Runs=np.array(runs)
```

```python
mycols=['b','g','r','c','m','y','k','r','g']
linestyles=['-','-','-','-','-','-','-','--','--']
```

```python
fname=x+'/'+str(tau[Runs[0]])+'daynokpp/grid.nc'
file2=netcdf.netcdf_file(fname,'r')
Zp1=file2.variables['Zp1']
Zp=Zp1[:]*1
dx=5000
dz=Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
dA=dz*5000
```

```python
filename=x+'/3daynokpp/VTprimebar.nc'
file2read = netcdf.NetCDFFile(filename,'r')
VT=VT=file2read.variables['VT']
VTzoneorig=np.mean(VT[:]*1,axis=2)
```

```python
for i in range(len(Runs)):
    fname=x+'/'+str(tau[Runs[i]])+'daynokpp/VTprimebar.nc'
    file2read = netcdf.NetCDFFile(fname,'r')
    VTbar=file2read.variables['VT']
    VTdA=np.zeros(shape(VTbar))
    for k in range(len(dz)):
        VTdA[k,:,:]=VTbar[k,:,:]*dA[k]
    EddyHF=1030*3985*(np.sum(np.sum(VTdA,axis=0),axis=1))/10**15
    EddyHF0=1030*3985*(np.sum(np.sum(VTdA[0:3,:,:],axis=0),axis=1))/10**15
    E0=plt.plot(EddyHF0,color=mycols[i],linestyle=linestyles[i],linewidth=2)
    #cbar=plt.colorbar(p, ticks=Psi_ticks, shrink=0.8)
    #ax.set_clim(-1,1)
    #cbar.ax.set_ylabel('$\psi \,\, (sv)$')
plt.title("Meridional Eddy Fluxes ("+str(Zp[3])+"m)",fontsize=20)
plt.ylabel("Heat Flux (PW)",fontsize=20)
plt.xlabel("Meridional Distance (km)",fontsize=20)
plt.tick_params(labelsize=16)
runnames=[]
for i in range(len(Runs)):
    runnames.append(tau[Runs[i]])
lgd=plt.legend(runnames, shadow=True,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,fontsize=20)
q=( os.path.expanduser('~')+"/Figures/Mobilis")
if not os.path.exists(q):
       os.makedirs(q)
plt.savefig(q+"/SurfaceEddy.png",bbox_extra_artists=(lgd,), bbox_inches='tight')
```
