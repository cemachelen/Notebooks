```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
import glob
import matplotlib.gridspec as gridspec
```

```python
# Define a class that forces representation of float to look a certain way
# This remove trailing zero so '1.0' becomes '1'
class nf(float):
     def __repr__(self):
         str = '%.1f' % (self.__float__(),)
         if str[-1]=='0':
             return '%.0f' % self.__float__()
         else:
             return '%.1f' % self.__float__()
```

```python
# Label levels with specially formatted floats
if plt.rcParams["text.usetex"]:
    fmt = r'%r $^oC$ %'
else:
    fmt = '%r $^oC$'
```

```python
Year='240-260all.nc'
```

```python
tau=['3','10','30','100','300','1000','3000','10000','Closed']
x='/noc/msm/scratch/students/hb1g13/Mobilis'
```

```python
check=0
runs=[]
for i in range(len(tau)):
    flist=x+'/'+str(tau[i])+'daynokpp/'+Year
    if not os.path.exists(flist):
       print 'WARNING: '+flist+' does not exist! (skipping this ta\
u...)'
       check+=0
    else:
       check+=1
       runs.append(i)
```

```python
gridfilepath=x+'/'+str(tau[0])+'daynokpp/'
file2=netcdf.netcdf_file(gridfilepath+'grid.nc','r')
Zp1=file2.variables['Zp1']
Zp=Zp1[:]*1
Z=file2.variables['Z']
Z=Z[:]
Y=file2.variables['Yp1']
Y=Y[:]
Yc=file2.variables['Y']
Yc=Yc[:]
```

```python
Runs=np.array(runs)
```

```python
fig=plt.figure(figsize=(12.5,12.5))
for i in range(len(Runs)):
    filename=x+'/'+str(tau[Runs[i]])+'daynokpp/'+Year
    file2read = netcdf.NetCDFFile(filename,'r')
    V=file2read.variables["VVEL"]
    V=V[:]
    Temp=file2read.variables['THETA']
    Temp=Temp[:]*1
    Tav=np.mean(Temp,axis=0)
    Tavlat=np.mean(Tav,axis=2)
    dx=Y[1]-Y[0] # Find Resolution
    Vtave=np.mean(V,axis = 0)
    Vtave[Vtave==0]=np.nan
    Vzone=np.nansum(Vtave,axis = 2)*dx
    dz=Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
    # Got rid of for loop here (much quicker!!)
    psi2=np.apply_along_axis(np.multiply,0,Vzone,dz)
    psi=np.cumsum(-psi2[::-1,:],axis=0)
    npad = ((0,1), (0,0))
    # Pad with zeros at bottom
    psi = np.pad(psi, pad_width=npad, mode='constant', constant_values=0)
    y =Y/1000
    Psi=psi/10**6 #Convert to Sv
    Psi[Psi>2.25]=2.25
    Q_levs = (np.arange(-26,26,2.4))
    Psi_levs = Q_levs / 10
    Q_ticks = np.arange(0,25,2.5)
    Psi_ticks = Q_ticks / 10
    ax = fig.add_subplot(3, 3, i+1)
    p=ax.contourf(Y/1000,Zp,Psi,Psi_levs,cmap=plt.cm.seismic) #Use b2r colourmap
    Q2_levs = (np.arange(0,8,1))
    q=ax.contour(Yc/1000,Z,Tavlat,Q2_levs,colors='k',linewidths = 2)
    q.levels = [nf(val) for val in q.levels ]
    plt.clabel(q,q.levels[::2], inline=1,fmt=fmt, fontsize=10)
    #clim(-1,1) # Put 0 to white
    #cbar=plt.colorbar(cf, ticks=Psi_ticks, shrink=0.8)
    ax.set_title(str(tau[Runs[i]])+' day')
    if str(tau[Runs[i]])=='Closed':
        ax.set_title(str(tau[Runs[i]]))
    ax.set_xlabel('Distance (km)')
    ax.set_ylabel('Depth (m)')
    #cbar=plt.colorbar(p, ticks=Psi_ticks, shrink=0.8)
    #ax.set_clim(-1,1)
    #cbar.ax.set_ylabel('$\psi \,\, (sv)$')
plt.tight_layout()
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar=fig.colorbar(p, cax=cax,ticks=Psi_ticks)
cbar.ax.set_ylabel('$\psi \,\, (sv)$')
q=( os.path.expanduser('~')+"/Figures/Mobilis")
if not os.path.exists(q):
       os.makedirs(q)
plt.savefig(q+"/MOCpanel.pdf",bbox_extra_artists=(cbar))
```
