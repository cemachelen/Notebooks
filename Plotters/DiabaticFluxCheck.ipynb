***
**Diabatic Eddy Flux Check**

\begin{equation}
\left. \psi_{res}\right|_{z=h_{ml}} \frac{\partial b_0}{\partial y} = \tilde{B_0}
\end{equation}

Where $\tilde{B_0}$ is the effective bouyancy. 

In M&R they suggest  $\tilde{B_0}$ is pretty much set up air-sea surface bouyancy fluxes i.e.

\begin{equation}
\tilde{B_0}= g \alpha / \rho_0 c_p)Q
\end{equation}

But actually it also has a contribution by diabatic eddy fluxes:

\begin{equation} \psi_{iso}(y)\frac{\partial \overline{b}}{\partial y} \simeq B + D \end{equation}

where

\begin{equation}
D=(1-\mu) \int_{-hml}^0 \frac{\partial \overline{v'b'}}{\partial y} \mathrm{d}z
\end{equation}

***

**The Check**

Previous scripts show that 

\begin{equation}
\left. \psi_{res}\right|_{z=h_{ml}} \frac{\partial b_0}{\partial y} \neq \tilde{B_0}
\end{equation}

I can take whats left and assume that makes up the diabatic component.

But I also have v'T' so i can calculate the actual diabatic component (with the assumption $\mu$ = 0 i.e. fully diapycnal). 

***

**The Script**

Fistly lets Calulate D from v'T'

1. \begin{equation} D= g \alpha \int_{-hml}^0 \frac{\partial \overline{v'T'}}{\partial y} \mathrm{d}z \end{equation}

2. Panel plot this to give a quick visual plot

3. Borrow the code from the Diabatic Eddy Pannel Script to Exract the calculated D, $\psi_{iso}$ and B and see if they match up. Is anything missing?

```python
def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
def regrid(Variable):
    Vc=(Variable[:,0:-1]+Variable[:,1::])/2
    return Vc
numba_regrid = autojit()(regrid)
numba_regrid.func_name = "numba_regrid"
```

```python
# Constant Parameters
alpha=2.000000E-04
G=-9.81
rho_0=1000
C_p=3985
Res=5000
Ly=2000e3
Lx=1000e3 
H=2985 
nz=30 
#Dimensions
nx=np.round(Lx/Res)
ny=np.round(Ly/Res)
dx=np.ones(nx)*Res
dy=np.ones(ny)*Res
x=(np.cumsum(dx)-dx/2)-Lx/2
y=(np.cumsum(dy)-dy/2)-Ly/2
[Y, X]=np.meshgrid(y,x) 
Q=10*(np.sin(Y*(3*pi/Ly)))
Q[:,ny-(np.round(ny/6)):ny]=0
Q=np.zeros(np.shape(Q))
Q[:,0:int(5*ny/36)]=-11*(np.sin(Y[:,0:int(5*ny/36)]*(pi/Y[1,int(5*ny/36)])))
Q[:,int(5*ny/36):int(20*ny/36)]=10*(np.sin((Y[:,int(5*ny/36):int(20*ny/36)]-Y[1,int(5*ny/36)])*(pi/(5*Y[1,int(15*ny/36)]))))
Q[:,int(20*ny/36):int(30*ny/36)]=-10.5*(np.sin((Y[:,int(20*ny/36):int(30*ny/36)]-Y[1,int(20*ny/36)])*(0.8*pi/(Y[1,int(10*ny/36)]))))
Q=Q+(-sum(Q)/(ny*nx)) 
Psipred=(-G*alpha/(rho_0*C_p))*-(Q[1,:])
```
