Eddy heat fluxes:

1$^{st}$ break down = simple mean and eddy components

$v\,=\, \overline{v}+v'$

$t\,=\,\overline{t}+t'$

Total Heat Flux: $C_p\,\rho \, \iint \overline{vt}\, dA$

Mean Heat Flux= $C_p\,\rho \, \iint \overline{v}\overline{t}\, dA$

Eddy Heat Flux= $C_p\,\rho \, \iint \overline{v't'}\, dA$

2$^{nd}$ break down = Mean Eddy and Standing Eddy components

```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
from pylab import *
from numba import autojit
import glob
from IPython.parallel import Client
```

```python
def find_nearest(array,value):
    idx = (np.abs(array-value)).argmin()
    return array[idx]
```

```python
Comp="Mobilis"
tau=['3','10','30','100','300','1000','3000','10000','Closed']
x="/noc/msm/scratch/students/hb1g13/"+Comp
```

```python
check=0
runs=[]
for i in range(len(tau)):
    flist=x+'/'+str(tau[i])+'daynokpp/HML.nc'
    if not os.path.exists(flist):
       print 'WARNING: '+flist+' does not exist! (skipping this ta\
u...)'
       check+=0
    else:
       check+=1
       runs.append(i)
Runs=np.array(runs)
```

```python
#Area matrix
#grd="/noc/msm/scratch/students/hb1g13/"+Comp+"/"+OP+"/grid/"
#os.chdir(grd)
file2=netcdf.netcdf_file(x+'/'+str(tau[Runs[0]])+'daynokpp/grid.nc','r')
Zp1=file2.variables['Zp1']
Zp=Zp1[:]*1
Z=file2.variables['Z']
Yp1=file2.variables['Yp1']
Yc=file2.variables['Y']
X=file2.variables['X']
dx=5000
dz=Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
dA=dz*5000
```

```python
fig=plt.figure(figsize=(12.5,10.5))
fig2=plt.figure(figsize=(12.5,10.5))
for i in range(len(Runs)):
    #Read in the VTtav files
    fileVTp = netcdf.NetCDFFile(x+'/'+str(tau[Runs[i]])+"daynokpp/VTprimebar.nc",'r') 
    VTbar=fileVTp.variables['VT']
    VTdA=np.zeros(shape(VTbar))
    for k in range(len(dz)):
        VTdA[k,:,:]=VTbar[k,:,:]*dA[k]
    EddyHF=1030*3985*(np.sum(np.sum(VTdA,axis=0),axis=1))/10**15
    #Read in the Mixed layer depth
    filehml = netcdf.NetCDFFile(x+'/'+str(tau[Runs[i]])+"daynokpp/HML.nc",'r') 
    hml=filehml.variables['MXLDEPTH']
    hml=np.squeeze(hml[:]*1)
    hmlav=np.mean(hml,axis=1)
    Eddymixed=np.zeros((len(Yc[:]*1),len(X[:]*1)))
    Permmixed=np.zeros((len(Yc[:]*1),len(X[:]*1)))
    for ix in range(len(X[:]*1)):
        for j in range(len(Yc[:]*1)):
            I=find_nearest(Z[:]*1,-hml[j,ix])
            b=nonzero(Z[:]*1==I)[0][0]
            Eddymixed[j,ix]=1030*3985*(np.sum(VTdA[0:b,j,ix],axis=0))/10**15
            Permmixed[j,ix]=Eddymixed[j,i]/-Z[b]
    Eddymlhf=np.sum(Eddymixed,axis=1)
    Permmlhf=np.sum(Permmixed,axis=1)
    ax2= fig2.add_subplot(3, 3, i+1)
    ax2.plot(Eddymlhf,'k')
    plt.ylim((-0.012,0))
    ax2.set_title('Mixed layer heat flux '+str(tau[Runs[i]])+' day')
    if str(tau[Runs[i]])=='Closed':
        ax2.set_title('Mixed layer heat flux'+str(tau[Runs[i]]))
    ax2.set_xlabel('Distance (km)')
    ax2.set_ylabel('Heat Flux PW')
plt.tight_layout()
q=( os.path.expanduser('~')+"/Figures/Mobilis")
if not os.path.exists(q):
       os.makedirs(q)
plt.savefig(q+"/Eddysmlpanel.png")
```
