***
**Bouyancy budget:**
\begin{equation}
J(\psi , b) = \frac{\partial B}{\partial z}-\frac{\partial}{\partial y}[(1- \mu ) \overline{v'b'}]
\end{equation}
So thats:
\begin{equation}
\frac{\partial \psi}{\partial y}\frac{\partial b}{\partial z}+\frac{\partial \psi}{\partial z}\frac{\partial b}{\partial y} = \frac{\partial B}{\partial z}-\frac{\partial}{\partial y}[(1- \mu ) \overline{v'b'}]
\end{equation}

For simplicity you can do this budget for the mixed layer and so you would assume no vertical bouyancy gradient.

My model doesn't have the maximums in the mixedlayer and deffining it is not so simple so there may well be parts where there is a vertical buoyancy gradient.

***

Integrate wrt to Z:

\begin{equation}
\frac{\partial \psi}{\partial y}b + \psi \frac{\partial b}{\partial y} = B - \int \frac{\partial}{\partial y}[(1- \mu ) \overline{v'b'}] dz 
\end{equation}

can split $\psi$ in to eddy and mean parts:

\begin{equation}
b(\frac{\partial \psi '}{\partial y} + \frac{\partial \overline{\psi} }{\partial y}) + \psi \frac{\partial b}{\partial y} = B - \int \frac{\partial}{\partial y}[(1- \mu ) \overline{v'b'}] dz
\end{equation}

Sub in wind and eddy fluxes to describe mean and eddy streamfunctions

\begin{equation}
b(\frac{\partial \tau}{\partial y} + \frac{\partial ^2 \overline{w'b'} }{\partial y^2}) + \psi \frac{\partial b}{\partial y} = B - \frac{\partial}{\partial y}[(1- \mu ) \overline{v'b'}]
\end{equation}

So now i've got ROC = Wind + bouyancy + eddies

Where I have local forcing and eddies modulated by chaging the northern boundary condition

Outside of the mixed layer: $\overline{w'b}$  = 0 so:

\begin{equation}
\psi = (B-\frac{b}{f \rho _0}\frac{\partial \tau}{\partial y})\frac{1}{b_y}
\end{equation}

Again ROC = ( Local Wind + bouancy ) * meridional bouyancy gradient (set by BN)
